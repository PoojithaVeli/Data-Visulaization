```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values_pie = [25, 40, 30, 45]

x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

x_bar = categories
values_bar = [10, 30, 20, 25]

data_hist = np.random.normal(size=100)

# Pie chart
fig_pie = px.pie(names=categories, values=values_pie, title='Pie Chart')

# Line chart
fig_line = px.line(x=x_line, y=y_line, title='Line Chart')

# Bar graph
fig_bar = px.bar(x=x_bar, y=values_bar, title='Bar Graph')

# Histogram
fig_hist = px.histogram(x=data_hist, nbins=20, title='Histogram')

# Display the dashboard
fig_pie.show()
fig_line.show()
fig_bar.show()
fig_hist.show()
```
